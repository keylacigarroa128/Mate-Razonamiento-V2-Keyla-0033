#### Problema 3
Enunciado: Un número al que se le resta 7 da como resultado 18.Paso 1: Explicación y Resolución
* Planteamiento: Definimos $x$ como el número desconocido. La frase "se le resta 7" se traduce como $x - 7$, y "da como resultado 18" es $= 18$.Ecuación:$$x - 7 = 18$$
* Despeje: Para dejar sola a la $x$, el 7 que está restando pasa al otro lado sumando:$x = 18 + 7$$x = 25$



```python
# Problema 3: Hallar el número
resultado_final = 18
resta = 7

# Despeje: x = resultado + resta
numero = resultado_final + resta

print(f"El número buscado es: {numero}")
```

#### Problema 4
Enunciado: El triple de un número más 4 es igual a 31.
* Paso 1: Explicación y ResoluciónPlanteamiento: "El triple de un número" es $3x$. "Más 4" es $+ 4$. "Igual a 31" es $= 31$.Ecuación:$$3x + 4 = 31$$
* Despeje:Primero quitamos lo que suma: $3x = 31 - 4$S
* implificamos: $3x = 27$El 3 que multiplica pasa dividiendo: $x = 27 / 3$$x = 9$

```python
# Problema 4: Triple de un número más 4
total = 31
suma = 4
multiplicador = 3

# Despeje: x = (total - suma) / multiplicador
x = (total - suma) / multiplicador

print(f"El número es: {int(x)}")
```

#### Problema 5
Enunciado: El largo mide 5 metros más que el ancho. Perímetro total = 50 metros.* Paso 1: Explicación y ResoluciónPlanteamiento: * Ancho = $x$Largo = $x + 5$Fórmula del perímetro del rectángulo: $P = 2(largo) + 2(ancho)$Ecuación:$$2(x + 5) + 2(x) = 50$$
* Despeje:Multiplicamos los paréntesis: $2x + 10 + 2x = 50$Sumamos las $x$: $4x + 10 = 50$Pasamos el 10 restando: $4x = 50 - 10 \rightarrow 4x = 40$
* Dividimos entre 4: $x = 40 / 4 \rightarrow$ $x = 10$ (Ancho)
* Resultados:Ancho: 10 metros.Largo: $10 + 5 =$ 15 metros.

```python
# Problema 5: Dimensiones del terreno
perimetro = 50
diferencia_largo = 5

# Ecuación simplificada: 4x + (2 * diferencia) = perimetro
# x = (perimetro - 2 * diferencia) / 4
ancho = (perimetro - (2 * diferencia_largo)) / 4
largo = ancho + diferencia_largo

print(f"El ancho mide: {ancho} metros")
print(f"El largo mide: {largo} metros")
```
